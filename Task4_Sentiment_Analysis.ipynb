```python
import pandas as pd
from textblob import TextBlob
data = pd.DataFrame({
    "Review": [
        "I love this product",
        "Worst experience ever",
        "The service was okay",
        "Amazing quality and fast delivery",
        "Not satisfied with the product",
        "Very happy with the purchase",
        "Poor customer support",
        "Average experience"
    ]
})
data["Review"] = data["Review"].str.lower()
def get_sentiment(text):
    polarity = TextBlob(text).sentiment.polarity
    if polarity > 0:
        return "Positive"
    elif polarity < 0:
        return "Negative"
    else:
        return "Neutral"

data["Sentiment"] = data["Review"].apply(get_sentiment)
data
data["Sentiment"].value_counts()

```

Insights:
1. Most reviews are positive, indicating good user satisfaction.
2. Negative reviews highlight areas for improvement.
3. Sentiment analysis helps understand customer opinions quickly.
4. NLP techniques convert unstructured text into meaningful insights.
Conclusion:
This task demonstrates how Natural Language Processing can be used to analyze textual data and classify sentiments effectively.
